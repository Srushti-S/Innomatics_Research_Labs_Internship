```python
import statistics

pop_std=statistics.pstdev(df['Salary'])
pop_std
```

```python
def z_score(sample_size, sample_mean, pop_mean, pop_std):
    numerator = sample_mean - pop_mean
    denomenator = pop_std / sample_size**0.5
    return numerator / denomenator

z = z_score(sample_size, sample_mean, pop_mean, pop_std)
```

```python
z
```

```python
def tscore(sample_mean, pop_mean, sample_std, sample_size):
    numerator = sample_mean - pop_mean
    denominator = sample_std/sample_size**0.5
    return numerator/denominator
    
sstd= np.std(df1.Salary)
t_sc = tscore(sample_mean,pop_mean, sstd,sample_size)
```

```python
t_sc
```

```python
#One-tail Test

confidence_level = 0.9
alpha = 1-0.95
t_critical = t.ppf(1-alpha, df=18)
t_critical
```

```python
#Two-tail Test

alpha = 1 - confidence_level
z_critical = norm.ppf(1 - alpha/2)
z_critical
```

```python
x_min = 230000
x_max = 350000

mean = pop_mean
std = sstd/(sample_size**0.5)

x = np.linspace(x_min, x_max, 100)
y = norm.pdf(x, mean, std)

plt.xlim(x_min, x_max)
plt.plot(x, y)


t_critical_left = pop_mean + (-t_critical * std)

x1 = np.linspace(x_min, t_critical_left, 100)
y1 = norm.pdf(x1, mean, std)
plt.fill_between(x1, y1, color='purple')

plt.scatter(sample_mean, 0)
plt.axvline(mean, color = 'orange')
plt.annotate("x_bar", (sample_mean, 0.02))
plt.show()
```

```python
x_min = 230000
x_max = 350000

mean = pop_mean
std = pop_std / sample_size**0.5

x = np.linspace(x_min, x_max, 100)
y = norm.pdf(x, mean, std)
plt.xlim(x_min, x_max)
plt.plot(x, y)

z_critical_left = pop_mean + (-z_critical * std)
z_critical_right = pop_mean + (z_critical * std)

x1 = np.linspace(x_min, z_critical_left, 100)
y1 = norm.pdf(x1, mean, std)
plt.fill_between(x1, y1, color='orange')

x2 = np.linspace(z_critical_right, x_max, 100)
y2 = norm.pdf(x2, mean, std)
plt.fill_between(x2, y2, color='green')

plt.scatter(sample_mean, 0)
plt.axvline(mean, color = 'black')
plt.annotate("x_bar", (sample_mean, 0.0005))
plt.figure(figsize=(20,10))
plt.show()
```

*Is there a relationship between gender and specialisation? (i.e. Does the preference of Specialisation depend on the Gender?)*

```python
df.groupby(by=['Gender','Specialization']).count()['ID']
```

```python
from scipy.stats import chi2
from scipy.stats import chi2_contingency
```

```python
x = np.linspace(0,100, 100)
y = chi2.pdf(x, df=4)
plt.plot(x, y)
```

```python
cont = pd.crosstab(df.Specialization,df.Gender)
cont
```

```python
chi2_contingency(cont)
```

```python
chi2_test_stat = chi2_contingency(cont)[0]
pval = chi2_contingency(cont)[1]
d = chi2_contingency(cont)[2]
```

```python
confidence_level = 0.95
alpha = 1 - confidence_level
chi2_critical = chi2.ppf(1 - alpha, d)
chi2_critical
```

```python
x_min = 0
x_max = 100

x = np.linspace(x_min, x_max, 100)
y = chi2.pdf(x, d)
plt.xlim(x_min, x_max)
plt.plot(x, y)

chi2_critical_right = chi2_critical

x1 = np.linspace(chi2_critical_right, x_max, 100)
y1 = chi2.pdf(x1, d)
plt.fill_between(x1, y1, color='blue')
```

```python
if(pval < alpha):
    print("Null Hypothesis is rejected")
else:
    print("Fail to reject the Null Hypothesis")
```

```python
if(chi2_test_stat > chi2_critical):
    print("Null Hypothesis is rejected")
else:
    print("Fail to reject the Null Hypothesis")
```
